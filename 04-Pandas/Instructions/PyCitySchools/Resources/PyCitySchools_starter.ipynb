```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load_csv = "Resources/schools_complete.csv"
student_data_to_load_csv = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load_csv)
student_data_df = pd.read_csv(student_data_to_load_csv)
```

```python
school_data_df.head(15)
```

```python
student_data_df.head()
```

```python
# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete_df.head(30)
```

```python
school_data_complete_df.columns
```

```python
reduced_school_data_complete_df = school_data_complete_df.loc[:, ['Student ID', 'school_name', 'budget',
                                             'math_score', 'reading_score']  ]

reduced_school_data_complete_df
```

```python
# count the number of unique schools in the DataFrame
school_count = len(reduced_school_data_complete_df["school_name"].unique())
school_count
```

```python
total_budget = (reduced_school_data_complete_df["budget"].unique().sum())
total_budget
```

```python
total_students = len(reduced_school_data_complete_df["Student ID"])
total_students
```

```python
# Calculate the average math score
average_math_score = reduced_school_data_complete_df["math_score"].sum() / total_students
average_math_score
```

```python
# Calculate the average reading score
average_reading_score = reduced_school_data_complete_df["reading_score"].sum() / total_students
average_reading_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)

passing_math_df = reduced_school_data_complete_df.loc[  reduced_school_data_complete_df["math_score"] >= 70     ]
passing_math_df.head(15)
```

```python
passing_math = len(passing_math_df)
passing_math
```

```python
percent_passing_math = (passing_math / total_students) * 100
percent_passing_math
```

```python
passing_reading_df = reduced_school_data_complete_df.loc[  reduced_school_data_complete_df["reading_score"] >= 70     ]
passing_reading_df.head(10)
```

```python
passing_reading = passing_reading_df["Student ID"].count()
percent_passing_reading = (passing_reading / total_students) * 100
percent_passing_reading
```

```python
Overall_passing = reduced_school_data_complete_df[  (reduced_school_data_complete_df['math_score'] >= 70)
                                                  & (reduced_school_data_complete_df['reading_score'] >= 70)    ]['Student ID'].count()/total_students*100
Overall_passing
```

```python
# Create a dataframe to hold the above results
summary_df = pd.DataFrame({
    "Total Schools": school_count,
    "Total Students": f"{total_students:,}",
    "Total Budget": f"${total_budget:,.2f}",
    "Average Math Score": f"{average_math_score:.6f}",
    "Average Reading Score": f"{average_reading_score:.5f}",
    "% Passing Math": f"{percent_passing_math:.6f}",
    "% Passing Reading": f"{percent_passing_reading:.6f}",
    "% Overall Passing": f"{Overall_passing: .6f}"
}, index=[0])
```

```python
# Create a dataframe to from the result dictionary
summary_df = pd.DataFrame(summary_df)
summary_df
```

```python
# Average Math Score

average_math_score = group_school_name['math_score'].mean()
average_math_score
```

```python
# Average Reading Score
average_reading_score = group_school_name['reading_score'].mean()
average_reading_score
```

```python
# % Passing Math
pass_math_percent = school_data_complete_df[  
    school_data_complete_df['math_score'] >= 70].groupby('school_name')['Student ID'].count()/total_student*100
pass_math_percent
```

```python
# % Passing Reading
pass_reading_percent = school_data_complete_df[  
    school_data_complete_df['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/total_student*100
pass_reading_percent
```

```python
# % Overall Passing (The percentage of students that passed math and reading.
overall_passing = school_data_complete_df[  
    (school_data_complete_df['reading_score'] >= 70) &
    (school_data_complete_df['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/total_student*100
overall_passing
```

```python
school_summary_df = pd.DataFrame({
    "School Type": school_type_df,
        "Total Students": total_student,
    "Total School Budget": total_school_budget,
    "Per Student Budget": budget_per_student,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    '% Passing Math': pass_math_percent,
    '% Passing Reading': pass_reading_percent,
    "% Overall Passing": overall_passing
})

#dropna? drop index? drop = true??
school_summary_df
```

```python
#formatting
school_summary_df.style.format({'Total Students': '{:}',
                             "Total School Budget": "${:,.2f}",
                          "Per Student Budget": "${:.2f}",
                             'Average Math Score': "{:6f}", 
                          'Average Reading Score': "{:6f}", 
                          "% Passing Math": "{:6f}", 
                          "% Passing Reading": "{:6f}"})
```

```python
# Create bins for segmenting and sorting data
bins = [0, 584, 629, 644, 675]

# Define name for each data that falls in each interval index(bin) 
spending_group_name = ['<$585', '$585-630', '$630-645', '$645-680']
```

```python
#group by spending
group_spending = school_summary_df.groupby(['spending_bins'])
group_spending.head()
```

```python
reading_by_spending = group_spending['Average Reading Score'].mean()
reading_by_spending
```
